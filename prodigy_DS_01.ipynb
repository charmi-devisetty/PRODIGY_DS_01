```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
csv_file = "/content/API_SP.POP.TOTL_DS2_en_csv_v2_73852.csv"
```

```python
df = pd.read_csv(csv_file, skiprows=4)
df_2023 = df[["Country Name", "2023"]].dropna()
df_2023["2023"] = df_2023["2023"] / 1_000_000
```

```python
# Plot histogram
plt.figure(figsize=(10, 6))
sns.histplot(df_2023["2023"], bins=20, kde=False, color='skyblue', edgecolor='black')
plt.title("Distribution of Country Populations in 2023", fontsize=14)
plt.xlabel("Population (millions)", fontsize=12)
plt.ylabel("Number of Countries", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
top10 = df_2023.sort_values(by="2023", ascending=False).head(10)
```

```python
# Plot bar chart for the top 10 countries
plt.figure(figsize=(12, 6))
sns.barplot(data=top10, x="Country Name", y="2023", palette="viridis")
plt.title("Top 10 Most Populous Countries in 2023", fontsize=14)
plt.xlabel("Country", fontsize=12)
plt.ylabel("Population (millions)", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
